```python
#!/usr/bin/env python
# coding: utf-8

# In[ ]:

import socket
import time
import numpy as np

ip = '192.168.10.1' # デフォルトのip
port = 8889 # ポート
socket = socket.socket(socket.AF_INET, socket.SOCK_DGRAM)
tello_address = (ip , port)
socket.sendto('command'.encode('utf-8'),tello_address)

#離陸のため'takeoff'コマンドを送る
#connect
time.sleep(7) #7秒休む
socket.sendto('takeoff'.encode('utf-8'),tello_address)

td = 0 #total distane 進んだ総距離を記録するためのいれもの
d = [] #distance 進んだ距離を記録するためのいれもの

for i in range(6): #6回繰り返す
    if td <= 700: #トータルの距離が700を超えるまで
        time.sleep(7)
        socket.sendto('forward 30'.encode('utf-8'),tello_address)
        time.sleep(7)
        d = np.append(d, 30) #dの配列に進んだ距離を追加
        socket.sendto('ccw 90'.encode('utf-8'),tello_address)     
        time.sleep(7)
        socket.sendto('left 120'.encode('utf-8'),tello_address)
        d = np.append(d, 120) #dの配列に進んだ距離を追加
        time.sleep(7)
        socket.sendto('up 50'.encode('utf-8'),tello_address)
        time.sleep(7)
        socket.sendto('forward 120'.encode('utf-8'),tello_address)
        d = np.append(d, 120) #dの配列に進んだ距離を追加
        time.sleep(7)
        socket.sendto('right 120'.encode('utf-8'),tello_address)
        d = np.append(d, 120) #dの配列に進んだ距離を追加
        time.sleep(7)
        socket.sendto('down 50'.encode('utf-8'),tello_address)
        time.sleep(7)
        socket.sendto('back 80'.encode('utf-8'),tello_address)
        d = np.append(d, 80) #dの配列に進んだ距離を追加 
        
    else: #トータルの距離が700を超えたら
        time.sleep(7)
        socket.sendto('forward 80'.encode('utf-8'),tello_address)
        d = np.append(d, 80) #dの配列に進んだ距離を追加
        time.sleep(7)
        socket.sendto('cw 80'.encode('utf-8'),tello_address)
    td = np.sum(d) #dの配列内の和を求める
    
#前に60センチ進む
time.sleep(10)
socket.sendto('forward 60'.encode('utf-8'),tello_address)
 
#90度曲がる
time.sleep(10)
socket.sendto('cw 90'.encode('utf-8'),tello_address)
    
#前に60センチ進む
time.sleep(10)
socket.sendto('forward 60'.encode('utf-8'),tello_address)
    
#90度曲がる
time.sleep(10)
socket.sendto('cw 90'.encode('utf-8'),tello_address)
    
#前に60センチ進む
time.sleep(10)
socket.sendto('forward 60'.encode('utf-8'),tello_address)
    
#90度曲がる
time.sleep(10)
socket.sendto('cw 90'.encode('utf-8'),tello_address)

#着陸のため'land'コマンドを送る
#land
time.sleep(7)
socket.sendto('land'.encode('utf-8'),tello_address)

#disconnect
time.sleep(5)
socket.close()
```
